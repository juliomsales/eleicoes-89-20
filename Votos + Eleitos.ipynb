```python
import pandas as pd
```

```python
eleitos = pd.read_csv('exports/eleitos.csv', sep=';', index_col=0)
partidos = pd.read_csv('exports/partidos.csv', sep=';')
```

```python
eleitos.drop('NM_CANDIDATO', axis=1, inplace=True)
```

```python
eleitos = eleitos.groupby(['DS_CARGO', 'SG_PARTIDO', 'SG_UF', 'ANO_ELEICAO']).sum().reset_index()
eleitos
```

```python
partidos['VOTOS'] = partidos['QT_VOTOS_NOMINAIS'] + partidos['QT_VOTOS_LEGENDA']
partidos.drop(['QT_VOTOS_NOMINAIS', 'QT_VOTOS_LEGENDA'], axis=1, inplace=True)
partidos
```

```python
mix = partidos.merge(eleitos, on=['DS_CARGO', 'SG_PARTIDO', 'SG_UF', 'ANO_ELEICAO'], how='left')
mix
```

```python
presidentes = eleitos[eleitos['DS_CARGO'] == 'PRESIDENTE']
```

```python
mix = mix.append(presidentes, ignore_index=True)
mix['NR_TURNO'].fillna(1, inplace=True)
mix = mix.append(presidentes, ignore_index=True)
mix['NR_TURNO'].fillna(2, inplace=True)
```

```python
mix.tail(20)
```

```python
mix.fillna(0, inplace=True)
```

```python
mix
```

```python
for column in mix.columns:
    if mix[column].dtype == 'float64':
        mix[column] = mix[column].astype(int)
```

```python
mix.to_csv('exports/votos_eleito.csv', sep=';', index=False)
```
